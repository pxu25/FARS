#### 5.For each state, what is the percentage of fatal accidents involved at least one drunk driver? If the column, DRUNK_DR, has a 0, then no drunk drivers were involved. Any number larger than 0 indicates the number of drunk drivers involved in the accident.

```python
print("State    ", "Percentage of fatal accidents involved at least one drunk driver")
for file in glob("accident_*.csv"):
    state_code = file[9:11]
    acc = pd.read_csv("accident_{}.csv".format(state_code))
    acc_fatal_one = acc[acc["FATALS"] & acc["DRUNK_DR"] > 0]
    acc_fatal_one_sum = sum(acc_fatal_one["FATALS"])
    acc_fatal_total = sum(acc["FATALS"])
    print(state["{}".format(state_code)],"{:.2%}".format(acc_fatal_one_sum/acc_fatal_total))
```

#### 6.For whole United States, what is the percentage of fatal accidents involved at least one drunk driver?

```python
acc_all = pd.DataFrame()
file_list = []
for file in glob("accident_*.csv"):
    acc = pd.read_csv(file,index_col=None, header=0)
    file_list.append(acc)
acc_all = pd.concat(file_list)

acc_FATALS = acc_all["FATALS"]
acc_DRUNK_one = acc_all["DRUNK_DR"] != 0
acc_all_fatal_one = acc_all[acc_FATALS & acc_DRUNK_one]
acc_all_fatal_one_sum = sum(acc_all_fatal_one["FATALS"])
acc_all_fatal_total = sum(acc_all["FATALS"])
print("{:.2%}".format(acc_all_fatal_one_sum/acc_all_fatal_total))
```

```python
collision_type = {
"0": "Not Collision with Motor Vehicle in Transport",
"1": "Front-to-Rear",
"2": "Front-to-Front",
"6": "Angle",
"7": "Sideswipe – Same Direction",
"8": "Sideswipe – Opposite Direction",
"9" : "Rear-to-Side",
"10" : "Rear-to-Rear",
"11" : "Other (End-Swipes and Others)",
"98" : "Not Reported",
"99" : "Unknown"
}

print("Collision_type", "                         Fatalities")
for key in collision_type:
    acc_all_key = acc_all[acc_all.MAN_COLL == int(key)]
    acc_all_key_sum = sum(acc_all_key['FATALS'])
    print("{:<30}{:>20}".format(collision_type.get(key), acc_all_key_sum))
```

#### 8.For each state, what is its fatal accident rate per 10,000 people? To calculate this, count the number of accidents in a state, divide by the state's 2016 population estimate from the nst-est2017-alldata.csv Census data file, and then multiply by 10000. Output the states' rates in order from highest to lowest.

```python
state = {
"01": "Alabama",    "31" : "Nebraska",
"02": "Alaska",     "32" : "Nevada",
"04": "Arizona",    "33" : "New Hampshire",
"05": "Arkansas",    "34" : "New Jersey",
"06": "California",   "35" : "New Mexico",
"08": "Colorado",     "36" : "New York",
"09": "Connecticut", "37": "North Carolina",
"10": "Delaware",    "38": "North Dakota",
"11": "District of Columbia", "39" : "Ohio",
"12": "Florida",    "40": "Oklahoma",
"13": "Georgia",    "41": "Oregon",
"15": "Hawaii",     "42": "Pennsylvania",
"16":  "Idaho",     "43": "Puerto Rico",
"17":  "Illinois",  "44": "Rhode Island",
"18": "Indiana",    "45": "South Carolina",
"19": "Iowa",       "46":  "South Dakota",
"20": "Kansas",     "47": "Tennessee",
"21": "Kentucky",   "48": "Texas",
"22":"Louisiana",   "49": "Utah",
"23": "Maine",      "50":"Vermont",
"24": "Maryland",   "52":"Virgin Islands (since 2004)",
"25": "Massachusetts", "51": "Virginia",
"26": "Michigan",    "53": "Washington",
"27": "Minnesota",   "54": "West Virginia",
"28": "Mississippi", "55":"Wisconsin",
"29": "Missouri",    "56": "Wyoming",
"30":"Montana"
}   

usa_pop = pd.read_csv("nst-est2017-alldata.csv")

state_name_list = []
state_fatal_list = []
for file in glob("accident_*.csv"):
    state_code = file[9:11]
    acc = pd.read_csv("accident_{}.csv".format(state_code))
    acc_fatal_sum = sum(acc["FATALS"])
    
    # State pop 2016
    state_name = state.get(state_code)
    state_name_list.append(state_name)
    state_pop = usa_pop[usa_pop['NAME'] == state_name]
    state_pop2016 = state_pop["POPESTIMATE2016"]
    fatal_rates = float(10000 * acc_fatal_sum / state_pop2016)
    state_fatal_list.append(str(fatal_rates))

# creat a DataFrame
df = pd.DataFrame({"state_name": state_name_list, "state_fatal_rate": state_fatal_list})
df.sort_values("state_fatal_rate", ascending=False)
```

#### 9.For each state, what is the rate of fatal accidents caused by drunk driving per 10,000 people? To calculate this, count the number of accidents in which a drunk driver was involved, divide by the state's 2016 population estimate from the nst-est2017-alldata.csv Census data file, and then multiply by 10000. Output the states' rates in order from highest to lowest.

```python
state_fatal_drunk1_list = []
for file in glob("accident_*.csv"):
    state_code = file[9:11]
    acc = pd.read_csv("accident_{}.csv".format(state_code))
    acc_drunk1 = acc[acc["DRUNK_DR"] != 0]
    acc_fatal_drunk1_sum = sum(acc_drunk1["FATALS"])
    
    # State pop 2016
    state_name = state.get(state_code)
    state_pop = usa_pop[usa_pop['NAME'] == state_name]
    state_pop2016 = state_pop["POPESTIMATE2016"]
    fatal_rates_drunk1 = float(10000 * acc_fatal_drunk1_sum / state_pop2016)
    state_fatal_drunk1_list.append(str(fatal_rates_drunk1))

# creat a DataFrame
df_drunk1 = pd.DataFrame({"state_name": state_name_list, "state_fatal_rate": state_fatal_drunk1_list})
df_drunk1.sort_values("state_fatal_rate", ascending=False)
```
